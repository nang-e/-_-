```python
units = ['g', '개', '/', 't', 'T', '장', '마리', 'ml', '팩', '인분', '줌', '컵', '봉지', '큰술', '알', '움큼', '주먹', '수저', '[고기밑간양념]',
       '쪽', '숟갈', '스푼', '흰자', '대', '토막', '꼬집', '톡톡', '티스푼', '적당량', '재료', '모', '숟가락', '[가니쉬]', '그릇', '그램', 
        '조각', '큰술', '줄', '호', '온탑', '뜨거운',  '비율', '채', '썬', 'ts', 'TS', '뿌리', '줄기', '봉', '[소스]', '[고기양념]', '구', 'C',
        '포기', '깐', '작은술', '손질전', '다진', '약간', '마음대로', '달랑', '것', '다진', '공기', '건', '정도', '[육수]', '묶음', '그램', '통', 
        '조금', '자반', '기타', '-', 'L', 'l', '[재료]', '준비물', '보관', '온도', '에어프라이어', '종류', '도구', '메인재료', '[메인재료]',
        '[주재료]', '[필수재료]', '[탈취재료]', '[반죽, 재료]', '[식재료]', '[세척재료]', '[양념]', '국자', 'cm', '청', '넉넉히', '또는', 'or', '캔', 
        '리터', 'b', '톨', '센치', '스쿱', '주걱', '덩이', '키로', '준비', '양념', '선택']
```

```python
ingreds = df['재료']
pre_ingred = []

for ingred in ingreds:
    validwords = []
    words = ingred.split(',')
    for word in words:
      delete = 0
      for unit in units:
          if unit in word:
            delete = 1
      if delete == 0:
        validwords.append(word)
    pre_ingred.append(validwords)
```

```python
df['전처리재료'] = pre_ingred
```
